```python
#LARS

lars = array(['02', '05', '02', '09', '11'])

#Lidman

lidman = array(['S115236', 'SGP8884'])
```

```python
#LARS 

for i in range(len(lars)): 
       
    num = lars[i]
    
    lars_data = loadtxt('./data/LARS'+num+'.txt')

    fig = figure(figsize=(16,9))
    
    ax = fig.add_subplot(1,1,1)
    ax.tick_params(axis='x', labelsize=25)
    ax.tick_params(axis='y', labelsize=25)

    x = lars_data[:,0]
    y = lars_data[:,1]
    
    lab = 'LARS'+num
    
    plot(x, y, label=lab, lw = 1.5, c='black')
    axhline(y=0, xmin=0, xmax=1, linestyle='--', c='black')

    legend(fontsize=28)
    
    xlabel('Velocity  (km s$^{-1}$)', fontsize=28)
    ylabel('Flux (mJy)', fontsize=28)
    
    #xlim(-500,2500)
    title('LARS'+num, fontsize=28)

    savefig('./data/LARS'+num+'.png')
```

```python
#Lidman

for i in range(len(lidman)): 
       
    num = lidman[i]
    
    lidman_data = loadtxt('./data/Lidman'+num+'.txt')

    fig = figure(figsize=(16,9))
    
    ax = fig.add_subplot(1,1,1)
    ax.tick_params(axis='x', labelsize=25)
    ax.tick_params(axis='y', labelsize=25)

    x = lidman_data[:,0]
    y = lidman_data[:,1]
    
    lab = 'Lidman'+num
    
    plot(x, y, label=lab, lw = 1.5, c='black')
    axhline(y=0, xmin=0, xmax=1, linestyle='--', c='black')

    legend(fontsize=28)
    
    xlabel('Velocity  (km s$^{-1}$)', fontsize=28)
    ylabel('Flux ($10^{-17}$s$^{-1}$cm$^{-2}\AA^{-1}$)', fontsize=28)
    
    #xlim(-500,2500)
    title('Lidman'+num, fontsize=28)

    savefig('./data/Lidman'+num+'.png')
```
