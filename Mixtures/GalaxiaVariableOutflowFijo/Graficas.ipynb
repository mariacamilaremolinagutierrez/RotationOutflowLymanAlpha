```python
#Numeros
num_tau=3
num_vmax=4
num_x=100

#Declaro arreglos
arreglo_vmax = array([0,100,200,300])
arreglo_tau=array([10**5,10**6,10**7])
angulo = 0
```

```python
data_path = "./data/res/rotout_"

espectros = zeros((num_x, num_vmax, num_tau))
x_center_analytic = zeros(100)
v_th=12.4
i_th=100
c=300000.0 #km/s
frecuencia_LyA_reposo=2.47*(10**15) #1/s

for j in arange(num_tau): 
    for i in arange(num_vmax):
    
        vmax=int(arreglo_vmax[i])
        logtau=int(log10(arreglo_tau[j]))
        
        filein = "%st%dv%da%d"%(data_path, logtau, vmax, angulo)
        
        data = loadtxt(filein+".dat")
        espectros[:,i,j] = data[:,1]
        x_center_analytic = data[:,0]
        delta_x = x_center_analytic[1] - x_center_analytic[0]
        norm_constant = sum(espectros[:,i,j]*delta_x)
        espectros[:,i,j] = data[:,1]/norm_constant
```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7, ax8), (ax9, ax10, ax11, ax12)) = plt.subplots(nrows=3, ncols=4, sharex=True, sharey=True)
fig.set_size_inches(12,9)
xlabel("$\mathrm{\lambda\ [Angstroms]}$")
#ylabel("$\mathrm{flux\ [arbitrary\ units]}$")

lambdas = c*1000/(frecuencia_LyA_reposo*(1 - x_center_analytic*v_th/c))

ax1.plot(lambdas, espectros[:,0,0]*i_th, color="black")
ax2.plot(lambdas, espectros[:,1,0]*i_th, color="black")
ax3.plot(lambdas, espectros[:,2,0]*i_th, color="black")
ax4.plot(lambdas, espectros[:,3,0]*i_th, color="black")
ax5.plot(lambdas, espectros[:,0,1]*i_th, color="black")
ax6.plot(lambdas, espectros[:,1,1]*i_th, color="black")
ax7.plot(lambdas, espectros[:,2,1]*i_th, color="black")
ax8.plot(lambdas, espectros[:,3,1]*i_th, color="black")
ax9.plot(lambdas, espectros[:,0,2]*i_th, color="black")
ax10.plot(lambdas, espectros[:,1,2]*i_th, color="black")
ax11.plot(lambdas, espectros[:,2,2]*i_th, color="black")
ax12.plot(lambdas, espectros[:,3,2]*i_th, color="black")

tight_layout()
savefig("rotout.png",format = 'png')
show()
close()
```
