```python
%pylab inline
from scipy.interpolate import interp1d
from read_LyaRT import wls
from os import listdir

ioff() #Apaga que se muestren las imagenes a menos que le de .show()
```

```python
#Datos

v_th=12.4
c=300000.0 #km/s
frecuencia_LyA_reposo=2.47*(10**15) #1/s

tau = 8 #tau = 10^8
v_gal = 300 #v_rot = 300 km/s

v_out = 500 #v_out = 500 km/s
logZ = -4.0 #log_metalicidad
```

```python
#Obtengo los datos de la galaxia

filename_galaxia = "t"+str(int(tau))+"v"+str(int(v_gal))+"a0.dat"

espectro_simulado=loadtxt("./data/rot/"+filename_galaxia)
velocidades_simuladas = espectro_simulado[:,0]*v_th
intensidades_simuladas = espectro_simulado[:,1]

#Paso de velocidades a longitud de onda
frecuencias_simuladas = frecuencia_LyA_reposo*(1 - velocidades_simuladas/c) #1/s
lambdas_simuladas = c*1000/frecuencias_simuladas #metros
lambdas_simuladas = lambdas_simuladas*10**10 #angstroms
```

```python
fig = figure()
plot(lambdas_simuladas,intensidades_simuladas)
xlabel("$\lambda$ (Angstrom)")
ylabel("Flux (Arbitrary Units)")
savefig("./rot_galaxia_inicial.png",format = 'png')
show()
close()  
```

```python
#Parametros del Outflow

filenames_TS = array(sorted(listdir("./data/out/ThinShell")))
filenames_W = array(sorted(listdir("./data/out/Wind")))
```

```python
#Primero: Wind

for filename in filenames_W:
    
    #Hallo wl_in y wl_out
    wl0, wl = wls("./data/out/Wind/"+filename)
    
    h=hist(wl0, bins=299)
    intensidades_planas, lambdas_planas = h[0], h[1]
    intensidad_plana_max = amax(intensidades_planas)
    
    #Hago que las intensidades queden del mismo orden que las de la distribucion plana
    intensidad_simulada_max = amax(intensidades_simuladas)
    intensidades_simuladas = intensidades_simuladas*intensidad_plana_max/intensidad_simulada_max

    #Hago la interpolacion de las simuladas    
    f_sim = interp1d(lambdas_simuladas, intensidades_simuladas)  
    
    #Evaluo w0 en esta nueva funcion
    intensidades_nuevas = f_sim(wl0)
        
    #Saco la figura -png y los datos del histograma .dat
        
    fig=figure()

    espectro_final=hist(wl, bins=100, weights=intensidades_nuevas, histtype='step')

    y=espectro_final[0]
    x=espectro_final[1]

    archivo = open("./data/rotout/Wind/rotout_"+filename+".dat",'w')

    for i in range (len(y)):
        archivo.write(str(x[i])+"\t"+str(y[i])+"\n")

    archivo.close()

    xlabel("$\lambda$ (Angstrom)")
    ylabel("Flux (Arbitrary Units)")
    xlim([])
    savefig("./data/img/Wind/rotout_"+filename+".png",format = 'png')
    show()
    close()  
```

```python
#Segundo: ThinShell

for filename in filenames_TS:
    
    #Hallo wl_in y wl_out
    wl0, wl = wls("./data/out/ThinShell/"+filename)
    
    h=hist(wl0, bins=299)
    intensidades_planas, lambdas_planas = h[0], h[1]
    intensidad_plana_max = amax(intensidades_planas)
    
    #Hago que las intensidades queden del mismo orden que las de la distribucion plana
    intensidad_simulada_max = amax(intensidades_simuladas)
    intensidades_simuladas = intensidades_simuladas*intensidad_plana_max/intensidad_simulada_max

    #Hago la interpolacion de las simuladas
    f_sim = interp1d(lambdas_simuladas, intensidades_simuladas)  
    
    #Evaluo w0 en esta nueva funcion
    intensidades_nuevas = f_sim(wl0)
        
    #Saco la figura -png y los datos del histograma .dat
        
    fig=figure()

    espectro_final=hist(wl, bins=100, weights=intensidades_nuevas, histtype='step')

    y=espectro_final[0]
    x=espectro_final[1]

    archivo = open("./data/rotout/ThinShell/rotout_"+filename+".dat",'w')

    for i in range (len(y)):
        archivo.write(str(x[i])+"\t"+str(y[i])+"\n")

    archivo.close()

    xlabel("$\lambda$ (Angstrom)")
    ylabel("Flux (Arbitrary Units)")
    savefig("./data/img/ThinShell/rotout_"+filename+".png",format = 'png')
    show()
    close()    
```
