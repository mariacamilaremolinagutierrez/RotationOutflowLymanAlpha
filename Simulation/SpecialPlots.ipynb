```python
%pylab inline
from pandas import *
from scipy.ndimage.filters import gaussian_filter1d
```

```python
#Importing dataframe

df = read_csv('./data/galaxyoutflow.csv')

v_gal = df['v_gal']
v_out = df['v_out']
lognh = df['lognh']
```

```python
#Changing lognh

#lognh_array = array([20.125,20.5625,21.25,22.25])
lognh_array = array([20.1875,21,22.1875])
df_lognh = df[(v_gal==100) & (v_out==100)]

fig = figure(figsize=(16,9))

ax = fig.add_subplot(1,1,1)
ax.tick_params(axis='x', labelsize=25)
ax.tick_params(axis='y', labelsize=25)

linestyles = ['-', '--', ':']
linewidths = [1, 2, 5]

for i in range(len(lognh_array)):
    
    ln = lognh_array[i]
    
    df_ln = df_lognh[lognh==ln]
    x = df_ln['x']
    y = df_ln['y']
    
    lab = "$\log{n_H} = $"+str(ln)
    
    plot(x, y, linestyles[i], label=lab, lw = linewidths[i], c='black')

legend(fontsize=28)
xlabel("Velocity  (km s$^{-1}$)", fontsize=28)
ylabel("Intensity (Arbitrary Units)", fontsize=28)
xlim(-500,2500)

savefig("./data/influences/inf_lognh.png")
```

```python
#Changing v_out with lognh1

vout_array = array([100,200,300])
df_vout = df[(lognh==20.125) & (v_gal==100)]

fig = figure(figsize=(16,9))

ax = fig.add_subplot(1,1,1)
ax.tick_params(axis='x', labelsize=25)
ax.tick_params(axis='y', labelsize=25)

linestyles = ['-', '--', ':']
linewidths = [1, 2, 5]

for i in range(len(vout_array)):
    
    vo = vout_array[i]
    
    df_vo = df_vout[v_out==vo]
    x = df_vo['x']
    y = df_vo['y']
    
    lab = "$v_{out} = $"+str(vo)+" km s$^{-1}$"
    
    plot(x, y, linestyles[i], label=lab, lw = linewidths[i], c='black')

legend(fontsize=28)
xlabel("Velocity  (km s$^{-1}$)", fontsize=28)
ylabel("Intensity (Arbitrary Units)", fontsize=28)
#xlim(-200,1200)
xlim(-300,2100)
ylim(0,7200)

savefig("./data/influences/inf_vout1.png")
```
