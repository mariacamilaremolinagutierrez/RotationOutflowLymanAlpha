```python
%pylab inline
from pandas import *
from scipy.ndimage.filters import gaussian_filter1d
```

```python
#Importing dataframe

df = read_csv('./data/galaxyoutflow.csv')

v_gal = df['v_gal']
v_out = df['v_out']
lognh = df['lognh']
```

```python
#tau = 2*10^5, vgal = 60 km/s, vout = 100 km/s, changing lognh

lognh_array = array([20.125, 20.625, 21.5, 22.1875])
df_lognh = df[(v_gal==60) & (v_out==100)]

fig = figure(figsize=(16,9))

ax = fig.add_subplot(1,1,1)
ax.tick_params(axis='x', labelsize=25)
ax.tick_params(axis='y', labelsize=25)

linestyles = ['-', '--', ':', '-.']
linewidths = [1, 2, 3, 2]

for i in range(len(lognh_array)):
    
    ln = lognh_array[i]
    
    df_ln = df_lognh[lognh==ln]
    x = df_ln['x']
    y = df_ln['y']
    
    lab = "$\log{n_H} = $"+str(ln)
    
    plot(x, y, linestyles[i], label=lab, lw = linewidths[i], c='black')

legend(fontsize=28)
xlabel("Velocity  (km s$^{-1}$)", fontsize=28)
ylabel("Intensity (Arbitrary Units)", fontsize=28)
xlim(-500,2500)
title("Galaxy #1", fontsize=28)

savefig("./data/influences/inf_lognh1.png")
```

```python
#tau = 2*10^6, vgal = 125 km/s, vout = 200 km/s, changing lognh

lognh_array = array([20, 20.4375, 21.375, 21.875])
df_lognh = df[(v_gal==125) & (v_out==200)]

fig = figure(figsize=(16,9))

ax = fig.add_subplot(1,1,1)
ax.tick_params(axis='x', labelsize=25)
ax.tick_params(axis='y', labelsize=25)

linestyles = ['-', '--', ':', '-.']
linewidths = [1, 2, 3, 2]

for i in range(len(lognh_array)):
    
    ln = lognh_array[i]
    
    df_ln = df_lognh[lognh==ln]
    x = df_ln['x']
    y = df_ln['y']
    
    lab = "$\log{n_H} = $"+str(ln)
    
    plot(x, y, linestyles[i], label=lab, lw = linewidths[i], c='black')

legend(fontsize=28)
xlabel("Velocity  (km s$^{-1}$)", fontsize=28)
ylabel("Intensity (Arbitrary Units)", fontsize=28)
xlim(-500,2500)
title("Galaxy #2", fontsize=28)

savefig("./data/influences/inf_lognh2.png")
```

```python
#tau = 2*10^5, vgal = 60 km/s, vout = 100 km/s, changing lognh

lognh_array = array([20.125, 20.625, 21.5, 22.1875])
df_lognh = df[(v_gal==60) & (v_out==100)]

fig = figure(figsize=(16,9))

ax = fig.add_subplot(1,1,1)
ax.tick_params(axis='x', labelsize=25)
ax.tick_params(axis='y', labelsize=25)

linestyles = ['-', '--', ':', '-.']
linewidths = [1, 2, 3, 2]

for i in range(len(lognh_array)):
    
    ln = lognh_array[i]
    
    df_ln = df_lognh[lognh==ln]
    x = df_ln['x']
    y = df_ln['y']
    y_gaussian = gaussian_filter1d(y,1.2)
    y_gaussian = y_gaussian*amax(y)/amax(y_gaussian)
    
    lab = "$\log{n_H} = $"+str(ln)
    
    plot(x, y_gaussian, linestyles[i], label=lab, lw = linewidths[i], c='black')

legend(fontsize=28)
xlabel("Velocity  (km s$^{-1}$)", fontsize=28)
ylabel("Intensity (Arbitrary Units)", fontsize=28)
xlim(-500,2500)
title("Galaxy #1", fontsize=28)

savefig("./data/influences/inf_lognh1_soft.png")
```
