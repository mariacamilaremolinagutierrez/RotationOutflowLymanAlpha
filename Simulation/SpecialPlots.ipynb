```python
%pylab inline
from pandas import *
from scipy.ndimage.filters import gaussian_filter1d
```

```python
#Importing dataframe

df = read_csv('./data/galaxyoutflow.csv')

v_gal = df['v_gal']
v_out = df['v_out']
lognh = df['lognh']
```

```python
#tau = 2*10^5, vgal = 0 km/s, vout = 100 km/s, changing lognh

lognh_array = array([20.3125,20.625,21.0625,21.75,22.3125])
df_lognh = df[(v_gal==0) & (v_out==100)]

fig = figure(figsize=(16,9))

ax = fig.add_subplot(1,1,1)
ax.tick_params(axis='x', labelsize=25)
ax.tick_params(axis='y', labelsize=25)

linestyles = ['-', '--', '-.', ':', '-']
linewidths = [1, 2, 2, 3, 1]

ln = lognh_array[0]
df_ln = df_lognh[lognh==ln]
x = array(df_ln['x'])
mitad = len(x)/2

for i in range(len(lognh_array)):
    
    ln = lognh_array[i]
    
    df_ln = df_lognh[lognh==ln]
    x = array(df_ln['x'])[:mitad]
    y = array(df_ln['y'])[:mitad]
    
    y_gaussian = gaussian_filter1d(y,1.2)
    y_gaussian = y_gaussian*amax(y)/amax(y_gaussian)
    
    lab = "$\log{n_H} = $"+str(ln)
    
    plot(x, y_gaussian, linestyles[i], label=lab, lw = linewidths[i], c='black')

legend(fontsize=28)
xlabel("Velocity  (km s$^{-1}$)", fontsize=28)
ylabel("Intensity (Arbitrary Units)", fontsize=28)
xlim(-400,2500)
ylim(0,5000)
title("Galaxy #1 (v = 0 km/s)", fontsize=28)

savefig("./data/influences/galaxy1_v0.png")
```
